```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

train = pd.read_csv("train.csv",encoding="utf-8")
train_df = train.drop(["PassengerId","Cabin","Name","Transported"],axis=1)
train_df
```

```python
y_train = train["Transported"]
y_train
```

```python
test = pd.read_csv("test.csv",encoding="utf-8")
test_df = test.drop(["PassengerId","Cabin","Name"],axis=1)
test_df
```

```python
data = pd.concat([train_df,test_df],ignore_index=True)
data
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
HomePlanet_most = data["HomePlanet"].dropna().value_counts()
HomePlanet_most
```

```python
data["HomePlanet"] = data["HomePlanet"].fillna("Earth")
```

```python
CryoSleep_most = data["CryoSleep"].dropna().value_counts()
CryoSleep_most
```

```python
data["CryoSleep"] = data["CryoSleep"].fillna("False")
```

```python
Destination_most = data["Destination"].dropna().value_counts()
Destination_most
```

```python
data["Destination"] = data["Destination"].fillna("TRAPPIST-1e")
```

```python
VIP_most = data["VIP"].dropna().value_counts()
VIP_most
```

```python
data["VIP"] = data["VIP"].fillna("False")
```

```python
data_na = data.isna().sum()
data_na_index = data_na[data_na>0].index
data_na_index
```

```python
data_na_median = data[data_na_index].median()
data_na_median
```

```python
data[data_na_index] = data[data_na_index].fillna(data_na_median)
```
